## Image Filters

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from matplotlib.pyplot import figure

img = cv2.imread('ai.jpeg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

img_noise = cv2.imread('ai_noise.png')
img_noise = cv2.cvtColor(img_noise, cv2.COLOR_BGR2RGB)

img_average = cv2.blur(img,(5,5))
img_gaussian = cv2.GaussianBlur(img,(5,5),0)
img_median = cv2.medianBlur(img_noise,5)
img_bilateral = cv2.bilateralFilter(img_noise,9,75,75)

plt.figure(figsize=(20, 20))
plt.subplot(3,3,1),plt.imshow(img,cmap = 'gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])

plt.subplot(3,3,2),plt.imshow(img_noise,cmap = 'gray')
plt.title('Noisy Image'), plt.xticks([]), plt.yticks([])

plt.subplot(3,3,3),plt.imshow(img_average,cmap = 'gray')
plt.title('Average Filter'), plt.xticks([]), plt.yticks([])

plt.subplot(3,3,4),plt.imshow(img_gaussian,cmap = 'gray')
plt.title('Gaussian Filter'), plt.xticks([]), plt.yticks([])

plt.subplot(3,3,5),plt.imshow(img_median,cmap = 'gray')
plt.title('Median Filter'), plt.xticks([]), plt.yticks([])

plt.subplot(3,3,6),plt.imshow(img_bilateral,cmap = 'gray')
plt.title('Bilateral Filter'), plt.xticks([]), plt.yticks([])
```

## Edge Detection

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
def EdgeDetection(image):
    img = cv2.imread(image)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    img_gaussian = cv2.GaussianBlur(gray,(3,3),0)
    
    #Laplacian Filter
    img_laplacian = cv2.Laplacian(img_gaussian,cv2.CV_64F)
    
    #Canny Edge Detection
    img_canny = cv2.Canny(img,100,200)
    
    #Sobel Operator
    img_sobelx = cv2.Sobel(img_gaussian,cv2.CV_8U,1,0,ksize=5)
    img_sobely = cv2.Sobel(img_gaussian,cv2.CV_8U,0,1,ksize=5)
    img_sobel = img_sobelx + img_sobely
    
    #Prewitt Operator
    kernelx = np.array([[1,1,1],[0,0,0],[-1,-1,-1]])
    kernely = np.array([[-1,0,1],[-1,0,1],[-1,0,1]])
    img_prewittx = cv2.filter2D(img_gaussian, -1, kernelx)
    img_prewitty = cv2.filter2D(img_gaussian, -1, kernely)
    img_prewitt = img_prewittx + img_prewitty
    
    plt.figure(figsize=(20, 20))
    plt.subplot(3,3,1),plt.imshow(img,cmap = 'gray')
    plt.title('Original Image'), plt.xticks([]), plt.yticks([])
    
    plt.subplot(3,3,2),plt.imshow(img_laplacian,cmap = 'gray')
    plt.title('Laplacian Filter'), plt.xticks([]), plt.yticks([])
    
    plt.subplot(3,3,3),plt.imshow(img_canny,cmap = 'gray')
    plt.title('Canny Edge Detection'), plt.xticks([]), plt.yticks([])
    
    plt.subplot(3,3,4),plt.imshow(img_sobelx,cmap = 'gray')
    plt.title('Sobel X'), plt.xticks([]), plt.yticks([])
    
    plt.subplot(3,3,5),plt.imshow(img_sobely,cmap = 'gray')
    plt.title('Sobel Y'), plt.xticks([]), plt.yticks([])
    
    plt.subplot(3,3,6),plt.imshow(img_sobel,cmap = 'gray')
    plt.title('Sobel'), plt.xticks([]), plt.yticks([])
    
    plt.subplot(3,3,7),plt.imshow(img_prewittx,cmap = 'gray')
    plt.title('Prewitt X'), plt.xticks([]), plt.yticks([])
    
    plt.subplot(3,3,8),plt.imshow(img_prewitty,cmap = 'gray')
    plt.title('Prewitt Y'), plt.xticks([]), plt.yticks([])
    
    plt.subplot(3,3,9),plt.imshow(img_prewitt,cmap = 'gray')
    plt.title('Prewitt'), plt.xticks([]), plt.yticks([])
    plt.show()
```

```python
EdgeDetection('ai.jpeg')
```
